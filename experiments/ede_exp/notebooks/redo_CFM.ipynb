```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, StratifiedKFold, KFold, StratifiedShuffleSplit
from sklearn.preprocessing import StandardScaler
import xgboost as xgb
from sklearn.metrics import precision_score, recall_score, jaccard_score, roc_auc_score, accuracy_score, classification_report, balanced_accuracy_score
from sklearn.metrics import confusion_matrix
# from sklearn.externals import joblib
import joblib
import os
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python

# data = np.random.rand(4, 4)
definitions = ["0", "cpu", "mem", "copy"]

data = np.array(
    [[450,2,0,0],
    [1,150,0,0],
    [2,0,150,0],
    [2,0,0,220]],
    dtype=np.float64
)
```

```python
plt.figure(figsize=(9, 6),
           dpi = 600)
ax = sns.heatmap(data, annot=True, yticklabels=definitions, xticklabels=definitions, annot_kws={"size": 10})
plt.title('Confusion Matrix Fold {}'.format(4), fontsize = 15) # title with fontsize 20
plt.xlabel('Ground Truth', fontsize = 15) # x-axis label with fontsize 15
plt.ylabel('Predictions', fontsize = 15) # y-axis label with fontsize 15
```
